```python
import os

# targetDir = r'C:\Users\Public\PartIIB project 2023_2024\Image collection without reaction\00AgNO3_mole_fraction\Outputs_Grayscale_Labelled_Images_STdevs\size_folder2\helical'
targetDir = r'C:\Users\Chappyyyyyy\Documents\size_folder'

def obtainSizeRange(targetDir, desiredInputNumber, num_bins):
    lstDir = os.listdir(targetDir)    
    numLst = [i/2 for i in range(num_bins)]
    sizeDict = {}
    for j in numLst:
      sizeDict[j] = 0
    for i in lstDir:
        try:
          # print(i)
          t_number = int(i.split("t-")[1].split("_")[0])
          lowerT = 200 - (200/desiredInputNumber) + 1
          if lowerT <= t_number <= 200:
              size = float(i[-17:-5])
              sizeRange = int(size*2)/2
              sizeDict[sizeRange] += 1
        except IndexError:
            pass

    print(sizeDict)
    print(dict(sorted(sizeDict.items())))
    print(dict(sorted(sizeDict.items(), key=lambda item: item[1])))
    return dict(sorted(sizeDict.items()))

sizeDict = obtainSizeRange(targetDir, 4, 68)
totalItems = sum(sizeDict.values())
print(totalItems)
totalItemsBelow28 = sum([sizeDict[i] for i in sizeDict if i < 28])
print(totalItemsBelow28)
sizeDict
```

```python
# obtain average number of contours per bin x_avg
sizeDict = obtainSizeRange(targetDir, 4, 68)
numLst = [i/2 for i in range(len(sizeDict)+1)]
print(numLst)
counter = 0
for j in numLst:
    try:
        counter += sizeDict[j]
    except KeyError:
        pass
x_avg = counter/(len(numLst))
x_avg
```

```python
def obtainDesiredDict(sizeDict, disregardFactor, trainProportion, upperFactor, x_avg):
    desiredDict = {}
    x_lower = x_avg*disregardFactor
    x_upper = (x_avg*upperFactor)/trainProportion
    numLst = [i/2 for i in range(len(sizeDict)+1)]
    
    for j in numLst:
        try:
            # x = number of contours per bin
            x = sizeDict[j]
            
            if x < x_lower:
                desiredDict[j] = 0
            
            if x_lower < x < x_upper:
                desiredDict[j] = int(trainProportion*x)
                
            if x > x_upper:
                desiredDict[j] = int(x_avg*upperFactor)
                
        except KeyError:
            pass
    return desiredDict

desiredDict = obtainDesiredDict(sizeDict, 0.5, 0.8, 8, x_avg)
totalItems = sum(desiredDict.values())
print(totalItems)
desiredDict
```
